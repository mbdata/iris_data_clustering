# Clustering with the Iris Dataset

## What is Clustering?

Clustering is a form of unsupervised machine learning which attempts to group points into classes (i.e. clusters) with similar properties and features. It is a common technique used in statistical analysis and there are many types of clustering with applications in various use cases. As it is an unsupervised form of learning, we will feed the data to the algorithm and allow it to make its own inferences, the prediction accuracy of which we will afterwards evaluate.

## Importing the Iris Dataset and Load Into DataFrame

There are a number of libraries that we will use for our clustering analysis, namely **sklearn**, a great tool that provides you with all the possible types of clustering algorithms, including the popular K-Means, and Hierarchical Clustering, both of which we will be demonstrating here. We will be using Pandas for fitting the data into DataFrames and matplotlib for visualization.

```python
#import necessary libraries
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
np.set_printoptions(precision=5, suppress=True)  # suppress scientific float notation
%matplotlib inline
```

The Iris dataset is included when you import sklearn, simply call it by storing it into a variable and load it into a DataFrame:

```python
#import iris dataset
iris = datasets.load_iris()
```

```python
#load data into dataframe
data = pd.DataFrame(iris.data)
```

```python
#assign column names
data.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
```

We would also like to store the labels separately because we will later compare them to the prediction results of our clustering.

```python
#load labels into a dataframe
labels = pd.DataFrame(iris.target)
labels.columns = ["Targets"]
```

```python
#take a look at the data
data.head()
```

## Visualize the Data

It is always important to see what our data looks like when plotted. We will plot two scatterplots, one looking at the Sepal values and the other at Petal values.

```python
#set figure size and colors
plt.figure(figsize=(12,5))
colormap = np.array(['purple', 'green', 'blue'])

#plot Sepal values
plt.subplot(1,2,1)
plt.scatter(data.Sepal_Length, data.Sepal_Width, c=colormap[labels.Targets], s=40)
plt.title('Sepal Length x Sepal Width')

#plot Petal values
plt.subplot(1,2,2)
plt.scatter(data.Petal_Length, data.Petal_Width, c=colormap[labels.Targets], s=40)
plt.title('Petal Length x Petal Width');
```

## K-Means Clustering

K-Means clustering attempts to group datapoints by computing distances between datapoints and cluster centres. It begins with us selecting the number of classes to use, after which it randomly initializes the center points. Then, each datapoint is classified by computing the distance between itself and each center point, assigning that datapoint to the nearest center. Afterwards, each group center is recomputed by taking the average of all the points belonging to that class. This process is repeated until the group centers do not move much with each new iteration. K-Means is a simple method that provides quick results (it runs in linear complexity O(n)), but it has the disadvantage that we must tell it how many clusters it should identify.

We use the K-Means function from sklearn to build our model by specifying the number of clusters, which here is 3 because we already know that the Iris data has three defined classes, and then we fit it to our data.

```python
#build model by choosing # of clusters
model = KMeans(n_clusters=3)

#create fit
model.fit(data);
```

Below are the prediction results from our K-Means model, conveniently displayed in a 1D array that we will use to compare to the actual results.

```python
#view clustering results
model.labels_
```

Visualising the prediction results, below, it appears that the algorithm did not know which labels in the prediction to assign to which classes. We will need to fix the labels.

```python
#visualize results (predicted vs actual)

#set plot size and colors
plt.figure(figsize=(12,5))
colormap = np.array(['purple', 'green', 'blue'])

#plot original Petal values
plt.subplot(1,2,1)
plt.scatter(data.Petal_Length, data.Petal_Width, c=colormap[labels.Targets], s=40)
plt.title('Actual')

#plot predicted Petal values
plt.subplot(1,2,2)
plt.scatter(data.Petal_Length, data.Petal_Width, c=colormap[model.labels_], s=40)
plt.title('Predicted');
```

Fixing the labels is simple; we need only to rearrange the colors.

```python
#fix labels
model.labels_ = np.choose(model.labels_, [1, 0, 2]).astype(np.int64)
```

Let's re-plot the results and compare:

```python
#visualize results with fixed labels (predicted vs actual)

#set plot size and colors
plt.figure(figsize=(12,5))
colormap = np.array(['purple', 'green', 'blue'])

#plot original
plt.subplot(1,2,1)
plt.scatter(data.Petal_Length, data.Petal_Width, c=colormap[labels.Targets], s=40)
plt.title('Actual')

#plot prediction
plt.subplot(1,2,2)
plt.scatter(data.Petal_Length, data.Petal_Width, c=colormap[model.labels_], s=40)
plt.title('Predicted');
```

As we can see, the K-Means classifier identified the one class at the bottom left correctly, but there is a discrepancy in the results for the top two clusters.

## Agglomerative Hierarchial Clustering

Agglomerative hierarchical clustering treats each datapoint as a single cluster at the beginning and then sequentially merges pairs of clusters. The algorithm stops when all clusters have been merged into a single cluster that contains all the data points. This hierarchical structure is represented as a dendrogram, a tree-like diagram.

We would first like to generate the linkage matrix. Using the 'ward' method (Ward variance minimization algorithm), we calculate the distance between newly-formed clusters, which will be used to build our dendrogram.

```python
# generate linkage matrix
linkage = linkage(data, 'ward')
```

We can also use the Cophenetic Correlation Coefficient to find the correlation between the actual pairwise distances of all our datapoints to those suggested by the hierarchical clustering. We would like a value that is as close to 1 as possible because this means the clustering well preserves the original distances.

```python
#import necessary libraries
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist

#calculate correlation coefficient
c, coph_dists = cophenet(linkage, pdist(data))
c
```

We get a Cophenetic Correlation Coefficient of 0.87, which means that the hierarchical clustering does a fairly good job of preserving the original distances.

## Visualize Results in a Dendrogram

We can now plot our dendrogram, which, as we mentioned, is a tree-like structure that shows us the order in which datapoints were merged and their pairwise distances.

```python
# construct dendrogram
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
dendrogram(
    linkage,
    leaf_rotation=90.,  
    leaf_font_size=8., 
)
plt.show()
```

The x-axis exhibits the indices of our datapoints whereas the y-axis shows the distances (calculated using the 'ward' method) at which the points were grouped together. The vertical lines tell us which clusters were being merged into a new cluster, and the horizontal lines represent those new cluster merges. The height of the vertical lines is the distance at which the clusters were from one another at the time of merging. As we can see, this algorithm identified 3 clusters without us telling it how many (like in K-Means).

Having a look at the iterations performed by the linkage algorithm, we see the distances at which datapoints were successively merged. We see the distances gradually increase as points become merged into fewer and fewer clusters and grow further apart. It is interesting to note that the largest jumps in distance occur at the last iterations, when our clusters are fully formed and split off from one another, as seen by the leap from 6.39941 to 12.3004 to 32.42801. Have a look at the above dendrogram and you will see these same distances on the y-axis at the splitting points of the 3 clusters.
